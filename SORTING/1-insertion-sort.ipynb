# Insertion sort (сортировка вставками)


Удобна для сортировки коротких последовательностей. Именно таким способом обычно сортируют карты: держа в левой руке уже упорядоченные карты и взяв в правой рукой очередную карту, мы вставляем её в нужное место, сравнивая с имеющимися и идя справа налево.

```python
import numpy as np
```

```python
def insertion_sort(arr):
    # Сортировку начинаем со второго элемента, 
    # т.к. считается, что первый элемент уже отсортирован
    for i in range(1, len(arr)):
        x = arr[i]
        # Сохраняем ссылку на индекс предыдущего элемента
        j = i - 1
        # Элементы отсортированного сегмента перемещаем вперёд,
        # если они больше (меньше) элемента для вставки
        while j >= 0 and arr[j] > x:
            arr[j+1] = arr[j]
            j -= 1
                
        # Вставляем элемент    
        arr[j+1] = x
        
    return arr
```

```python
A = np.random.randint(-1000, 1000, 10000)
np_sorted = np.sort(A)
```

```python
%%time
our_sorted = insertion_sort(A)
```

```python
(np_sorted == our_sorted).sum()
```
